```python
def Preprocessing(df):
    
    #created Unsaturated Fats Column
    df['Unsaturated_fats'] = df['fatContent'] - df['saturatedFatContent']
    
    #Converted sodium into grams
    df['sodiumContent'] = (df['sodiumContent']/1000)
    
    #Considering Nutritional facts in terms of Calories
    df['saturatedFatContent_c'] = df['saturatedFatContent']*9
    df['Unsaturated_fats_c'] = df['Unsaturated_fats']*9
    df['carbohydrateContent_c'] = df['carbohydrateContent']*4
    df['proteinContent_c'] = df['proteinContent']*4
    df['fatContent_c'] = df['fatContent']*9
    
    #Calculating Persentage of each nutritional fact wrt. Calories of food
    
    df['saturatedFatContent%'] = (df['saturatedFatContent_c']/df['calories'])*100
    df['Unsaturated_fats%'] = (df['Unsaturated_fats_c']/df['calories'])*100
    df['carbohydrateContent%'] = (df['carbohydrateContent_c']/df['calories'])*100
    df['proteinContent%'] = (df['proteinContent_c']/df['calories'])*100
    df['fatContent%'] = (df['fatContent_c']/ df['calories'])*100
    
    #dropiing Unnecessary columns
    
    df.drop(['saturatedFatContent_c','Unsaturated_fats_c','carbohydrateContent_c','proteinContent_c','fatContent_c'],
          axis=1, inplace=True )
    
    #Converting sodium into daily value persentage
    
    df['sodiumContent%'] = (df['sodiumContent']/ 2.3)*100
    
    #Converting sugar into daily value persentage
    
    df['sugarContent%'] = (df['sugarContent']/ 50)*100
    
    
    
    
    return df
    
    
    
    
    
    
    
    
    
    
    
```

```python
def calculate_bmr_tdee(age, weight, height, gender, activeness, BMI):
    
    activity_multiplier = {"Sedentary": 1.2, "Lightly Active": 1.375, "Moderately Active": 1.55, "Very Active": 1.725,"Extra Active": 1.9}

    
    if gender == "Male":
        bmr = round(5 + (10 * weight) + (6.25 * height) - (5 * age), ndigits=2)
        tdee = round(bmr * activity_multiplier.get(activeness), ndigits=2)
        print("\nUser's BMR - {}kcal & TDEE - {}kcal".format(bmr, tdee))
    else:
        bmr = round((10 * weight) + (6.25 * height) - (5 * age) - 161, ndigits=2)
        tdee = round(bmr * activity_multiplier.get(activeness), ndigits=2)
        print("\nUser's BMR - {}kcal & TDEE - {}kcal".format(bmr, tdee))
    
    
    if BMI < 18.5:
        final_tdee = tdee + 850
        print("Final TDEE - {}kcal".format(final_tdee))
        
    elif (BMI>=18.5) & (BMI<24.9):
        final_tdee = tdee
        print("Final TDEE - {}kcal".format(final_tdee))
        
    else:
        final_tdee = tdee - 500
        print("Final TDEE - {}kcal".format(final_tdee))
        
    return bmr, final_tdee    
```

```python
from math import sqrt

# calculate the Euclidean distance between two vectors
def euclidean_distance(row1, row2):
	distance = 0.0
	for i in range(len(row1)-1):
		distance += (row1[i] - row2[i])**2
	return sqrt(distance)

# Locate the most similar neighbors
def get_neighbors(train, test_row, num_neighbors):
	distances = list()
	for train_row in train:
		dist = euclidean_distance(test_row, train_row)
		distances.append((train_row, dist))
	distances.sort(key=lambda tup: tup[1])
	neighbors = list()
	for i in range(num_neighbors):
		neighbors.append(distances[i][0])
	return neighbors
```

```python
from sklearn.decomposition import PCA

def Inputs(data, Recipe_input):
    
    arr = data.to_numpy()
#     pca = PCA()
#     pca = PCA(n_components=2)
#     new_data = pca.fit_transform(arr)
    new_data = arr.tolist()
    
    input_Recipe = Recipe_input.to_numpy()
#     input_Breakfast = pca.transform(input_Breakfast)
    input_Recipe = input_Recipe.tolist()
    
    return new_data, input_Recipe
```

```python
def Breakfast_Recommendation(final_tdee, df_recommend):
    index = []
    Calorie = 0.325* final_tdee   #700  range is 650-750
    
#     calorie_range = range(int(Calorie*0.95) , int(Calorie*1.05))
    a = int(Calorie*0.95)
    b = int(Calorie*1.05)
# sample = df_recommend['calories'].sample(n=1)



    sample1 = df_recommend['calories'].sample(n=1)
    sample2 = df_recommend['calories'].sample(n=2)


    if (sample1.values[0]>a) & (sample1.values[0]<b):
        index_sample1 = sample1.index.tolist()
        index.append(index_sample1[0])
        
    
    
    
    elif ((sample2.values[0] + sample2.values[1])>a) & ((sample2.values[0] + sample2.values[1])<b):
        index_sample2 = sample2.index.tolist()
        index.append(index_sample2[0])
        index.append(index_sample2[1])
        
    
    
    else:
        
        Sum = 0
        
        while True :
            
            sample3 = df_recommend['calories'].sample(n=3)



            Sum = sample3.values[0]+ sample3.values[1] + sample3.values[2]
        
            if (Sum>a) & (Sum<b):

                index_sample3 = sample3.index.tolist()
                index.append(index_sample3[0])
                index.append(index_sample3[1])
                index.append(index_sample3[2])
                break
                
    return index
    
           
```

```python
def Recommendation(df, age, weight, height, gender, activeness, BMI, health_issues):
    
    df = Preprocessing(df)
    
    df_Breakfast = df.loc[(df['Course']==1)]
    df_Lunch = df.loc[(df['Lunch_Course']==1)]
    df_Dinner = df.loc[(df['Dinner_Course']==1)]
    
    
    
    bmr, final_tdee  = calculate_bmr_tdee(age, weight, height, gender, activeness, BMI)
    data_input_Breakfast = df_Breakfast[["sodiumContent%",'sugarContent%','saturatedFatContent%','carbohydrateContent%','proteinContent%','Unsaturated_fats%']]
    data_input_Lunch = df_Lunch[["sodiumContent%",'sugarContent%','saturatedFatContent%','carbohydrateContent%','proteinContent%','Unsaturated_fats%']]
    data_input_Dinner = df_Dinner[["sodiumContent%",'sugarContent%','saturatedFatContent%','carbohydrateContent%','proteinContent%','Unsaturated_fats%']]
    
    Recommended_Breakfast_Recipe = Breakfast_Recipe(health_issues)
    Recommended_Lunch_Recipe = Lunch_Recipe(health_issues)
    Recommended_Dinner_Recipe = Dinner_Recipe(health_issues)
    
    Breakfast_new_data, Breakfast_input_Recipes = Inputs(data_input_Breakfast, Recommended_Breakfast_Recipe)
    Lunch_new_data, Lunch_input_Recipes = Inputs(data_input_Lunch, Recommended_Lunch_Recipe)
    Dinner_new_data, Dinner_input_Recipes = Inputs(data_input_Dinner, Recommended_Dinner_Recipe)
    
    
    
    Breakfast_Recipe_index = []
    Breakfast_Recipe_Neighbor =[]

    neighbors = get_neighbors(Breakfast_new_data, Breakfast_input_Recipes[0], 1000)

    for neighbor in neighbors:
        Breakfast_Recipe_Neighbor.append(neighbor)
    
    for results in Breakfast_Recipe_Neighbor:
        index = Breakfast_new_data.index(results)
        Breakfast_Recipe_index.append(index)
    
    
    Breakfast_df_Recipe = df.iloc[Breakfast_Recipe_index]
    
#     --------------------------------------------------------------------------------------
    
    
    Lunch_Recipe_index = []
    Lunch_Recipe_Neighbor =[]

    neighbors = get_neighbors(Lunch_new_data, Lunch_input_Recipes[0], 1000)

    for neighbor in neighbors:
        Lunch_Recipe_Neighbor.append(neighbor)
    
    for results in Lunch_Recipe_Neighbor:
        index = Lunch_new_data.index(results)
        Lunch_Recipe_index.append(index)
    
    
    Lunch_df_Recipe = df.iloc[Lunch_Recipe_index]
    
#     -----------------------------------------------------------------------------------------
    
    
    Dinner_Recipe_index = []
    Dinner_Recipe_Neighbor =[]

    neighbors = get_neighbors(Dinner_new_data, Dinner_input_Recipes[0], 1000)

    for neighbor in neighbors:
        Dinner_Recipe_Neighbor.append(neighbor)
    
    for results in Dinner_Recipe_Neighbor:
        index = Dinner_new_data.index(results)
        Dinner_Recipe_index.append(index)
    
    
    Dinner_df_Recipe = df.iloc[Dinner_Recipe_index]
    
#     ---------------------------------------------------------------------------------------------
    
    
    
    
    
    
    
    Breakfast_data = Breakfast_df_Recipe.loc[(Breakfast_df_Recipe['sodiumContent%']<25) & (Breakfast_df_Recipe['sugarContent%']<25) & (Breakfast_df_Recipe['fatContent%']<35)]
    Lunch_data = Lunch_df_Recipe.loc[(Lunch_df_Recipe['sodiumContent%']<25) & (Lunch_df_Recipe['sugarContent%']<25) & (Lunch_df_Recipe['fatContent%']<35)]
    Dinner_data = Dinner_df_Recipe.loc[(Dinner_df_Recipe['sodiumContent%']<25) & (Dinner_df_Recipe['sugarContent%']<25) & (Dinner_df_Recipe['fatContent%']<35)]
    
    
    
    BR = Breakfast_Recommendation(final_tdee, Breakfast_data )
    LR = Lunch_Recommendation(final_tdee, Lunch_data )
    DR = Dinner_Recommendation(final_tdee, Dinner_data )
    
    return  Breakfast_data, Lunch_data, Dinner_data, BR, LR, DR

    
    
    
    
    
    
    
    
```
